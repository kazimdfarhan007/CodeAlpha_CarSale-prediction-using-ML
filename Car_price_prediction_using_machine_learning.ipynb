####Data Preprocessing

```python
date_time=datetime.datetime.now()
data['Age']=date_time.year-data['Year']
data.head()
```

```python
data.drop('Year',axis=1,inplace=True)
```

```python
data.head()
```

####Outlier Removal

```python
sns.boxplot(data['Selling_Price']).set(ylabel='Selling_Price')
```

```python
sorted(data['Selling_Price'])
```

```python
data=data[~(data['Selling_Price']>=33.0)]
data.shape
```

####Encoding the Categorical columns

```python
data.head(1)
```

```python
data1=data
```

```python
data1['Fuel_Type'].unique()
```

```python
# Import label encoder
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'Fuel_Type'.
data1['Fuel_Type']= label_encoder.fit_transform(data1['Fuel_Type'])
data1['Fuel_Type'].unique()

```

```python
data1['Selling_type'].unique()
```

```python
# Encode labels in column 'Selling_type'.
data1['Selling_type']= label_encoder.fit_transform(data1['Selling_type'])
data1['Selling_type'].unique()
data1['Selling_type'].unique()
```

```python
data1['Transmission'].unique()
```

```python
data1['Transmission']= label_encoder.fit_transform(data1['Transmission'])
data1['Transmission'].unique()
data1['Transmission'].unique()
```

####Store Feature Matrix In X and Response(Target) In Vector y

```python
X=data1.drop(['Car_Name','Selling_Price'],axis=1)
y=data1['Selling_Price']
```

```python
y
```

####Splitting The Dataset Into The Training Set And Test Set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

####Impring Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)

```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
```

```python
print(score1,score2,score3)
```

```python
final_data = pd.DataFrame({'Models':['lr','rf','xgb'],
             "R2_SCORE":[score1,score2,score3]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'],y=final_data['R2_SCORE'],data=final_data)
```
